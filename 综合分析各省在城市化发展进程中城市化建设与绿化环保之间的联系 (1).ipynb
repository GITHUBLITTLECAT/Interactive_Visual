# 综合分析各省在城市化发展进程中城市化建设与绿化环保之间的联系
## 项目摘要
城市建设发展是推动我们城市化的重要途径，近年来，我国各省份城市建成区不断扩大。但随着城市化的逐步推进，在加速经济发展的同时，与带来了一系列生态环境问题。本项目旨在通过综合分析2014-2017年各省份城市建成区、地区生产总值、各省份绿化覆盖率等数据，对各省份的城市化发展与环保绿化工作做一个初步的评估

* 数据来源:国家数据


```python
from pyecharts import options as opts
from pyecharts.charts import Bar, Grid, Line,Scatter


def grid_vertical() -> Grid:
    bar = (
        Bar()
        .add_xaxis(a)
        .add_yaxis("2014年各省份GDP", a4)
        .set_global_opts(title_opts=opts.TitleOpts(title="2014年各省份GDP"),xaxis_opts=opts.AxisOpts(name_rotate=60,name="省份",axislabel_opts={"rotate":45}))
    )
    
    line = (
        Line()
        .add_xaxis(a)
        .add_yaxis("2014年城市建成区面积", b4)
        .set_global_opts(
            title_opts=opts.TitleOpts(title="2014年城市建成区面积", pos_top="48%"),
            legend_opts=opts.LegendOpts(pos_top="48%"),xaxis_opts=opts.AxisOpts(name_rotate=60,name="省份",axislabel_opts={"rotate":45})
        )
    )

    grid = (
        Grid()
        .add(bar, grid_opts=opts.GridOpts(pos_bottom="60%"))
        .add(line, grid_opts=opts.GridOpts(pos_top="60%"))
    )
    return grid

grid_vertical().render_notebook()
```

```python
import pandas as pd
import numpy as np
from pyecharts.charts import Geo
from pyecharts import options as opts
from pyecharts.globals import ChartType, SymbolType,ThemeType
from pyecharts.charts import Bar,Tab,Line,Map,Timeline

def timeline_map() -> Timeline:
    tl = Timeline()
    for i in range(2014,2018):
        map0 = (
            Map()
            .add(
                "城市绿化覆盖率",list(zip(list(df.province),list(df["{}年".format(i)]))),"china",is_map_symbol_show = False
            )
            .set_global_opts(
                title_opts=opts.TitleOpts(title="{}年城市绿化覆盖率".format(i),subtitle="",
                                          subtitle_textstyle_opts=opts.TextStyleOpts(color="red",font_size=18,font_style="italic")),
                visualmap_opts=opts.VisualMapOpts(min_= 30,max_=50)
            
            )
        )
        tl.add(map0,"{}年".format(i))
    return tl
timeline_map().render_notebook()
```

# 总结结论
结合各省人均碳排和各省单位GDP碳排数据柱状图，我们可以初步探究各省在城市化发展进程中城市化建设与绿化环保的联系。
* 总体来说，东部沿海经济较发达的省份，不仅城市建成区多，在城市绿化覆盖率上也要高于经济相对落后的省份。
* 例如经济相对发达的省份广东省虽然城市在城市建城区面积很大，但是城市绿化覆盖率、人均碳排、单位GDP碳排均属于全国比较优秀的水平，说明广东省在城市化经济发展的过程中即注重城市建设，同时也有注意城市绿化和环保工作。
* 经济相对落后的省份山西，城市建成区面积不多，城市绿化覆盖率也一般，单位GDP碳排却是全国第一。说明山西省在城市化发展上还需更加注重经济发展的质量。降低碳排的同时，提升城市绿化覆盖率，改善城市环境。
